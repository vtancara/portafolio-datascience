## **Ejercicio 1:** Atención de Emergencias
Desde un registro de quejas generadas desde la línea de atención a emergencias, se requiere identificar a las zonas afectadas por deslizamientos. Debido a la repetición de ubicaciones, se requiere que escriba un programa Python para eliminar todos los duplicados de una lista determinada de cadenas y devolver una lista de cadenas únicas. Utilice el tipo de datos sets de Python para resolver el ejercicio. Puede crear un conjunto de quejas escritas de ejemplo.

```python
class Queja:
    """Clase para representar una queja por deslizamiento."""

    def __init__(self, id_queja, zona, fecha, descripcion):
        """
        Inicializa una nueva queja por deslizamiento.
        """
        self.id_queja = id_queja
        self.zona = zona
        self.fecha = fecha
        self.descripcion = descripcion

    def __str__(self):
        """Devuelve una representación en cadena de la queja."""
        return f"Queja #{self.id_queja}: Zona {self.zona} - {self.fecha} - {self.descripcion}"

class RegistroQuejas:
    """Clase para gestionar un registro de quejas por deslizamientos."""

    def __init__(self):
        """Inicializa un nuevo registro de quejas vacío."""
        self.quejas = []

    def agregar_queja(self, queja):
        """
        Agrega una queja al registro.
        """
        self.quejas.append(queja)

    def obtener_zonas_unicas(self):
        """
        Obtiene una lista de zonas únicas afectadas por deslizamientos.
        """
        # Utiliza un set para eliminar duplicados
        zonas_unicas = set()

        for queja in self.quejas:
            zonas_unicas.add(queja.zona)

        # Convierte el set de vuelta a una lista
        return list(zonas_unicas)

    def mostrar_quejas_por_zona(self, zona):
        """
        Muestra todas las quejas de una zona específica.
        """
        return [queja for queja in self.quejas if queja.zona == zona]

def main():
    """Función principal para demostrar el funcionamiento del programa."""
    # Crear un nuevo registro de quejas
    registro = RegistroQuejas()

    # Crear algunas quejas de ejemplo
    quejas_ejemplo = [
        Queja("001", "Zona Norte", "2025-04-01", "Deslizamiento en ladera que afecta 5 viviendas"),
        Queja("002", "Zona Sur", "2025-04-01", "Pequeño deslizamiento en carretera principal"),
        Queja("003", "Zona Norte", "2025-04-02", "Nuevo deslizamiento cerca del anterior"),
        Queja("004", "Zona Este", "2025-04-02", "Deslizamiento que bloquea acceso a comunidad"),
        Queja("005", "Zona Oeste", "2025-04-03", "Deslizamiento menor sin daños estructurales"),
        Queja("006", "Zona Sur", "2025-04-03", "Deslizamiento que afecta línea eléctrica"),
        Queja("007", "Zona Norte", "2025-04-04", "Ampliación de deslizamiento previo"),
        Queja("008", "Zona Este", "2025-04-04", "Segundo deslizamiento en la misma área")
    ]

    # Agregar las quejas al registro
    for queja in quejas_ejemplo:
        registro.agregar_queja(queja)

    # Obtener la lista de zonas únicas
    zonas_unicas = registro.obtener_zonas_unicas()

    # Mostrar resultados
    print("Zonas afectadas por deslizamientos:")
    for i, zona in enumerate(zonas_unicas, 1):
        print(f"{i}. {zona}")

    print("\nDetalle de quejas por zona:")
    for zona in zonas_unicas:
        print(f"\n--- {zona} ---")
        quejas_zona = registro.mostrar_quejas_por_zona(zona)
        for queja in quejas_zona:
            print(f"  * {queja}")

if __name__ == "__main__":
    main()
```

## **Ejercicio 2:** Tesis de Postgrado

Mediante uso de estructuras de datos, crear un programa que le permita a la biblioteca de la UAGRM SOE gestionar las tesis de postgrado. El programa debe ser capaz de realizar las siguientes operaciones:
-  Agregar una nueva tesis: El usuario debe poder ingresar los detalles de una nueva tesis, como título, autor, tutor, carrera, grado académico, año, y cantidad de copias físicas disponible.
-  Buscar una tesis: El usuario debe poder buscar una tesis por título, autor o carrera. El programa debe mostrar los detalles de la tesis si se encuentra.
-  Actualizar la información de una tesis: El usuario debe poder modificar los detalles de una tesis existente.
-  Eliminar una tesis: El usuario debe poder eliminar una tesis del inventario.
-  Listar todas las tesis: El programa debe poder mostrar una lista de todas las tesis en el inventario, junto con sus detalles.

Implementar el programa considerando que algunas carreras en la universidad permiten que dos tesistas puedan trabajar en una misma tesis y que la mayoría de las carreras asignan dos tutores a un mismo proyecto (uno conocido como tutor y otro como relator). También se sugiere implementar un menú de opciones para que el usuario pueda interactuar fácilmente con el programa.


```python
class Tesis:

    def __init__(self, id, titulo, autores, tutores, carrera, grado, anio, copias):
        self.id = id
        self.titulo = titulo
        self.autores = autores
        self.tutores = tutores
        self.carrera = carrera
        self.grado = grado
        self.anio = anio
        self.copias = copias

    def mostrar(self):
        print(f"\nID: {self.id}")
        print(f"Título: {self.titulo}")
        print(f"Autores: {', '.join(self.autores)}")
        print(f"Tutores: {', '.join(self.tutores)}")
        print(f"Carrera: {self.carrera}")
        print(f"Grado: {self.grado}")
        print(f"Año: {self.anio}")
        print(f"Copias disponibles: {self.copias}")


class GestorTesis:

    def __init__(self):
        self.tesis = []
        self.siguiente_id = 1

    def agregar_tesis(self, titulo, autores, tutores, carrera, grado, anio, copias):
        nueva_tesis = Tesis(
            self.siguiente_id, titulo, autores, tutores, carrera, grado, anio, copias
        )
        self.tesis.append(nueva_tesis)
        self.siguiente_id += 1
        print(f"Tesis agregada con ID: {nueva_tesis.id}")
        return nueva_tesis

    def buscar_tesis(self, termino):

        resultados = []
        termino = termino.lower()

        for t in self.tesis:

            if termino in t.titulo.lower():
                resultados.append(t)
                continue

            for autor in t.autores:
                if termino in autor.lower():
                    resultados.append(t)
                    break

            if termino in t.carrera.lower():
                resultados.append(t)

        return resultados

    def actualizar_tesis(self, id, titulo=None, autores=None, tutores=None,
                         carrera=None, grado=None, anio=None, copias=None):

        for t in self.tesis:
            if t.id == id:
                if titulo: t.titulo = titulo
                if autores: t.autores = autores
                if tutores: t.tutores = tutores
                if carrera: t.carrera = carrera
                if grado: t.grado = grado
                if anio: t.anio = anio
                if copias is not None: t.copias = copias
                print(f"Tesis con ID {id} actualizada correctamente")
                return True

        print(f"No se encontró tesis con ID {id}")
        return False

    def eliminar_tesis(self, id):

        for i, t in enumerate(self.tesis):
            if t.id == id:
                del self.tesis[i]
                print(f"Tesis con ID {id} eliminada correctamente")
                return True

        print(f"No se encontró tesis con ID {id}")
        return False

    def listar_tesis(self):

        if not self.tesis:
            print("No hay tesis registradas")
            return

        print(f"\nLista de Tesis ({len(self.tesis)}):")
        for t in self.tesis:
            t.mostrar()

def ejecutar_ejemplo():
    gestor = GestorTesis()

    gestor.agregar_tesis(
        "Análisis de algoritmos de machine learning para diagnóstico médico",
        ["Juan Pérez", "María García"],
        ["Dr. Carlos López", "Dra. Ana Martínez"],
        "Ciencia de Datos",
        "Maestría",
        2023,
        3
    )

    gestor.agregar_tesis(
        "Implementación de redes neuronales en procesamiento de imágenes",
        ["Pedro Rodríguez"],
        ["Dr. Luis Sánchez", "Dr. Roberto Torres"],
        "Inteligencia Artificial",
        "Doctorado",
        2024,
        2
    )

    gestor.listar_tesis()

    print("\nBúsqueda de tesis con 'machine':")
    resultados = gestor.buscar_tesis("machine")
    for t in resultados:
        t.mostrar()

    gestor.actualizar_tesis(1, copias=5)

    print("\nDespués de actualizar:")
    gestor.listar_tesis()

    gestor.eliminar_tesis(2)

    print("\nDespués de eliminar:")
    gestor.listar_tesis()


ejecutar_ejemplo()
```

```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def mostrar(self):
        print(f"Nombre: {self.nombre}, Edad: {self.edad}")


class Cliente(Persona):
    def __init__(self, nombre, edad, telefono_de_contacto):
        super().__init__(nombre, edad)
        self.telefono_de_contacto = telefono_de_contacto

    def mostrar(self):
        super().mostrar()
        print(f"Teléfono de contacto: {self.telefono_de_contacto}")


class Empleado(Persona):
    def __init__(self, nombre, edad, sueldo_bruto):
        super().__init__(nombre, edad)
        self.sueldo_bruto = sueldo_bruto

    def mostrar(self):
        super().mostrar()
        print(f"Sueldo bruto: {self.sueldo_bruto}")

    def calcular_salario_neto(self):
        # Ejemplo simple con 20% de descuento
        return self.sueldo_bruto * 0.8


class Directivo(Empleado):
    def __init__(self, nombre, edad, sueldo_bruto, categoria):
        super().__init__(nombre, edad, sueldo_bruto)
        self.categoria = categoria
        self.subordinados = []

    def mostrar(self):
        super().mostrar()
        print(f"Categoría: {self.categoria}")
        print("Subordinados:")
        for sub in self.subordinados:
            sub.mostrar()

    def agregar_subordinado(self, empleado):
        self.subordinados.append(empleado)


class Empresa:
    def __init__(self, nombre):
        self.nombre = nombre
        self.empleados = []
        self.clientes = []

    def agregar_empleado(self, empleado):
        self.empleados.append(empleado)

    def agregar_cliente(self, cliente):
        self.clientes.append(cliente)

    def mostrar(self):
        print(f"Empresa: {self.nombre}")
        print("\nEmpleados:")
        for emp in self.empleados:
            emp.mostrar()
            print()
        print("Clientes:")
        for cli in self.clientes:
            cli.mostrar()
            print()
```

```python
def main():
    # Crear clientes
    cliente1 = Cliente("Ana Pérez", 30, "71234567")
    cliente2 = Cliente("Luis Torres", 45, "76543210")

    # Crear empleados
    empleado1 = Empleado("Carlos Gómez", 28, 5000)
    empleado2 = Empleado("Lucía Rojas", 35, 5500)

    # Crear directivo y asignar subordinados
    directivo1 = Directivo("María Fernández", 40, 8000, "Gerente General")
    directivo1.agregar_subordinado(empleado1)
    directivo1.agregar_subordinado(empleado2)

    # Crear empresa y agregar empleados y clientes
    empresa = Empresa("TechSoluciones")
    empresa.agregar_empleado(empleado1)
    empresa.agregar_empleado(empleado2)
    empresa.agregar_empleado(directivo1)
    empresa.agregar_cliente(cliente1)
    empresa.agregar_cliente(cliente2)

    # Mostrar toda la información
    empresa.mostrar()


# Ejecutar el programa
if __name__ == "__main__":
    main()
```

## **Ejercicio 4:** La biblioteca

Diseñe un diagrama de clases para el siguiente caso e implemente su respectiva POO con Python:

> Una biblioteca universitaria contiene libros. Hay múltiples copias de algunos libros. Algunos libros están disponibles para préstamo de una semana solamente, pero todos los demás libros pueden ser prestados por tres semanas generalmente. Los miembros de la biblioteca pueden tomar prestados hasta cuatro libros a la vez, y los miembros del personal pueden tomar prestamos de hasta doce libros. Los miembros de la biblioteca y los miembros del personal deben devolver luego los libros a la biblioteca como cualquier otro usuario.

Se sugiere considerar a los sustantivos como posibles objetos y a las acciones como posibles funciones.

```python
class Libro:
    def __init__(self, codigo, titulo, rapido=False):
        self.codigo = codigo
        self.titulo = titulo
        self.rapido = rapido
        self.disponible = True

    def mostrar(self):
        estado = "Disponible" if self.disponible else "Prestado"
        tiempo = "1 semana" if self.rapido else "3 semanas"
        print(f"{self.codigo} - {self.titulo} ({estado}, {tiempo})")


class Usuario:
    def __init__(self, nombre, limite=4):
        self.nombre = nombre
        self.limite = limite
        self.prestamos = []

    def prestar(self, libro):
        if len(self.prestamos) >= self.limite:
            print(f"{self.nombre} ya alcanzó el límite.")
            return
        if not libro.disponible:
            print(f"'{libro.titulo}' no está disponible.")
            return
        libro.disponible = False
        self.prestamos.append(libro)
        print(f"{self.nombre} prestó '{libro.titulo}'.")

    def devolver(self, libro):
        if libro in self.prestamos:
            libro.disponible = True
            self.prestamos.remove(libro)
            print(f"{self.nombre} devolvió '{libro.titulo}'.")

    def ver_prestamos(self):
        print(f"{self.nombre} tiene estos libros:")
        for l in self.prestamos:
            print(f"- {l.titulo}")


class Empleado(Usuario):
    def __init__(self, nombre):
        super().__init__(nombre, limite=12)


class Biblioteca:
    def __init__(self):
        self.libros = []

    def agregar(self, libro):
        self.libros.append(libro)

    def ver_catalogo(self):
        for l in self.libros:
            l.mostrar()
```

```python
def main():
    b = Biblioteca()
    l1 = Libro("001", "El Quijote")
    l2 = Libro("002", "Introducción a Python")
    l3 = Libro("003", "Matemáticas I", rapido=True)

    b.agregar(l1)
    b.agregar(l2)
    b.agregar(l3)

    u1 = Usuario("Laura")
    e1 = Empleado("Profe Ruiz")

    b.ver_catalogo()
    u1.prestar(l1)
    u1.prestar(l3)
    e1.prestar(l2)

    print()
    u1.ver_prestamos()
    print()
    b.ver_catalogo()


if __name__ == "__main__":
    main()
```

```python
class Usuario:
    def __init__(self, primer_nombre, otros_nombres, apellido_paterno, apellido_materno, ci):
        self.primer_nombre = primer_nombre.strip().lower()
        self.otros_nombres = otros_nombres.strip().lower()
        self.apellido_paterno = apellido_paterno.strip().lower()
        self.apellido_materno = apellido_materno.strip().lower()
        self.ci = ci.strip()
        self.correo = self.generar_correo()

    def generar_correo(self):
        iniciales_nombre = self.primer_nombre[:2]
        apellido = self.apellido_paterno
        ci_prefix = self.ci[:2]
        return f"{iniciales_nombre}{apellido}{ci_prefix}@cesma.edu.bo"

    def get_tipo(self):
        return "usuario"

    def __str__(self):
        return self.correo


class Docente(Usuario):
    def get_tipo(self):
        return "docente"


class Estudiante(Usuario):
    def get_tipo(self):
        return "estudiante"  # Se muestra como "usuario" según ejemplo


class GestorCuentas:
    def __init__(self):
        self.lista_usuarios = []

    def correo_existente(self, correo):
        return any(u.correo == correo for u in self.lista_usuarios)

    def crear_usuario(self, tipo, primer_nombre, otros_nombres, apellido_paterno, apellido_materno, ci):
        if tipo.lower() == "docente":
            nuevo_usuario = Docente(primer_nombre, otros_nombres, apellido_paterno, apellido_materno, ci)
        elif tipo.lower() == "estudiante":
            nuevo_usuario = Estudiante(primer_nombre, otros_nombres, apellido_paterno, apellido_materno, ci)
        else:
            raise ValueError("Tipo no válido")

        # Verificación de duplicado exacto
        if self.correo_existente(nuevo_usuario.correo):
            print("La cuenta ya existe para otro docente. Comuniquese con el administrador.")
            return None

        self.lista_usuarios.append(nuevo_usuario)
        print(f"El correo electrónico del {nuevo_usuario.get_tipo()} es: {nuevo_usuario.correo}")
        return nuevo_usuario

    def mostrar_usuarios(self):
        print("\nCuentas registradas:")
        for usuario in self.lista_usuarios:
            print(usuario)


gestor = GestorCuentas()

gestor.crear_usuario("estudiante", "Ana", "", "Mamani", "Lopez", "9112345")
gestor.crear_usuario("docente", "Luis", "", "Quispe", "Zeballos", "9011223")
gestor.crear_usuario("docente", "Carlos", "", "Limachi", "Callisaya", "8944221")
gestor.crear_usuario("estudiante", "Monica", "", "Tapia", "Gomez", "7399999")

# Intentar duplicar (mismo correo que ya existe)
gestor.crear_usuario("docente", "Ana", "", "Mamani", "Lopez", "9112345")

# Mostrar cuentas registradas
gestor.mostrar_usuarios()
```

## **Ejercicio 6:** Nuevas becas
Una fundación academica internacional ha decidido estimular a todos los estudiantes de una universidad local mediante la asignación de becas mensuales, para esto se tomarán en consideración los siguientes criterios:

Para alumnos mayores de 18 años con promedio mayor o igual a 90, la beca será de Bs.2000.00; con promedio mayor o igual a 75, de Bs.1000.00; para los promedios menores de 75 pero mayores o iguales a 60, de Bs.500.00; a los demás se les enviará una carta de invitación incitándolos a que estudien más en el próximo semestre.

A los alumnos de 18 años o menores de esta edad, con promedios mayores o iguales a 90, se les dará Bs.3000; con promedios menores a 90 pero mayores o iguales a 80, Bs.2000; para los alumnos con promedios menores a 80 pero mayores o iguales a 60, se les dará Bs.100, y a los alumnos que tengan promedios menores a 60 se les enviará carta de invitación.

En el caso de que un alumno se encuentre actualmente trabajando, se le asignara una beca de Bs.1000. Este sera un caso similar para alumnos que traben como asistentes de investigación en la fundación.

Se requiere la creación de una base de datos de estudiantes y su respectiva abstracción con POO para facilitar la programación del caso. Diseñe el diagrama de clases respectivo.  



### DIAGRAMA:

```python
class Beca:
    def __init__(self, monto=0.0, tipo="invitación"):
        self.monto = monto
        self.tipo = tipo  # 'económica' o 'invitación'

    def __str__(self):
        if self.tipo == "económica":
            return f"Beca económica asignada: Bs.{self.monto:.2f}"
        else:
            return "Carta de invitación: ¡Estudia más el próximo semestre!"


class Estudiante:
    def __init__(self, nombre, edad, promedio, trabaja=False, asistente=False):
        self.nombre = nombre
        self.edad = edad
        self.promedio = promedio
        self.trabaja = trabaja
        self.asistente = asistente
        self.beca = None  # Se asigna más adelante

    def calcular_beca(self):
        # Reglas especiales por trabajo o asistencia
        if self.trabaja or self.asistente:
            self.beca = Beca(1000, "económica")
            return

        # Mayores de 18
        if self.edad > 18:
            if self.promedio >= 90:
                self.beca = Beca(2000, "económica")
            elif self.promedio >= 75:
                self.beca = Beca(1000, "económica")
            elif self.promedio >= 60:
                self.beca = Beca(500, "económica")
            else:
                self.beca = Beca()
        # 18 o menores
        else:
            if self.promedio >= 90:
                self.beca = Beca(3000, "económica")
            elif self.promedio >= 80:
                self.beca = Beca(2000, "económica")
            elif self.promedio >= 60:
                self.beca = Beca(100, "económica")
            else:
                self.beca = Beca()

    def get_info(self):
        return f"{self.nombre} (Edad: {self.edad}, Promedio: {self.promedio}) -> {self.beca}"


class GestorBecas:
    def __init__(self):
        self.estudiantes = []

    def agregar_estudiante(self, estudiante):
        self.estudiantes.append(estudiante)

    def procesar_becas(self):
        for estudiante in self.estudiantes:
            estudiante.calcular_beca()

    def mostrar_resultados(self):
        print("\n=== Resultados de asignación de becas ===")
        for est in self.estudiantes:
            print(est.get_info())


# Crear el sistema gestor
gestor = GestorBecas()

# Agregar estudiantes (nombre, edad, promedio, trabaja, asistente)
gestor.agregar_estudiante(Estudiante("Carlos Pérez", 19, 92))
gestor.agregar_estudiante(Estudiante("Ana Torres", 18, 85))
gestor.agregar_estudiante(Estudiante("Luis Quispe", 21, 59))
gestor.agregar_estudiante(Estudiante("Marta Mamani", 17, 78))
gestor.agregar_estudiante(Estudiante("Pablo Flores", 20, 88, trabaja=True))
gestor.agregar_estudiante(Estudiante("Lucía Choque", 22, 75, asistente=True))

# Calcular y mostrar becas
gestor.procesar_becas()
gestor.mostrar_resultados()
```

## **Ejercicio 7.** Nuevos bonos

El gobierno de la Argentina ha decidido lanzar un bono mensual que se pagara tres veces a los médicos y al personal hospitalario de ese país. También, se cuenta con la información de la edad, antigüedad (años de trabajo) y el numero de dependientes de n trabajadores hospitalarios. Por cada dependiente, se cuenta con un registro que define la la fecha de nacimiento y edad del dependiente, y si estudia actualmente. Este bono se calcula en base a las siguientes condiciones:
	Si la antigüedad < 5 años:  		       10 % del Salario del empleado
	Si la antigüedad >=5 pero es menor a 10:     15 % del Salario del empleado
	Si la antigüedad >=10 pero es menor a 15:   25 % del Salario del empleado
	Si la antigüedad >= 15:	  		        50 % del Salario del empleado

Además de esto, el bono se incrementa en un 2% por familiar dependiente al trabajador y un 1% adicional si el dependiente estudia. Se requiere calcular el monto total que el gobierno tendrá que pagar al finalizar el tercer mes. Simular los datos de los trabajadores usando valores aleatorios y modelar el caso con POO. Implementar un las clases y funciones respectivas con Python.



```python
from datetime import date

class Dependiente:

    def __init__(self, nombre, anio_nacimiento, estudia):

        self.nombre = nombre
        self.anio_nacimiento = anio_nacimiento
        self.estudia = estudia

    def calcular_edad(self):

        anio_actual = date.today().year
        return anio_actual - self.anio_nacimiento

    def __str__(self):

        estado = "Estudia" if self.estudia else "No estudia"
        edad = self.calcular_edad()
        return f"{self.nombre} - {edad} años - {estado}"


class Empleado:

    def __init__(self, nombre, salario, antiguedad):

        self.nombre = nombre
        self.salario = salario
        self.antiguedad = antiguedad
        self.dependientes = []

    def agregar_dependiente(self, dependiente):
        self.dependientes.append(dependiente)

    def calcular_porcentaje_antiguedad(self):

        if self.antiguedad < 5:
            return 0.10
        elif self.antiguedad < 10:
            return 0.15
        elif self.antiguedad < 15:
            return 0.25
        else:
            return 0.50

    def calcular_porcentaje_dependientes(self):
        porcentaje_adicional = 0.0

        for dependiente in self.dependientes:
            porcentaje_adicional += 0.02

            if dependiente.estudia:
                porcentaje_adicional += 0.01

        return porcentaje_adicional

    def calcular_bono(self):

        porcentaje_antiguedad = self.calcular_porcentaje_antiguedad()
        porcentaje_dependientes = self.calcular_porcentaje_dependientes()

        porcentaje_total = porcentaje_antiguedad + porcentaje_dependientes

        bono = self.salario * porcentaje_total

        return bono


    def mostrar_detalles_bono(self):

        porcentaje_antiguedad = self.calcular_porcentaje_antiguedad()
        porcentaje_dependientes = self.calcular_porcentaje_dependientes()
        porcentaje_total = porcentaje_antiguedad + porcentaje_dependientes

        bono_mensual = self.salario * porcentaje_total
        bono_total = bono_mensual * 3


        print(f"\nDetalles del bono para {self.nombre}:")
        print(f"Salario: ${self.salario:.2f}")
        print(f"Antigüedad: {self.antiguedad} años")
        print(f"Dependientes: {len(self.dependientes)}")

        print("\nCálculo de porcentaje:")
        print(f"  - Por antigüedad: {porcentaje_antiguedad*100:.1f}%")
        print(f"  - Por dependientes: {porcentaje_dependientes*100:.1f}%")
        print(f"  - Total: {porcentaje_total*100:.1f}%")

        print(f"\nBono mensual: ${bono_mensual:.2f}")
        print(f"Bono total (3 meses): ${bono_total:.2f}")

    def __str__(self):
        return f"{self.nombre} - Salario: ${self.salario:.2f} - Antigüedad: {self.antiguedad} años"

class SistemaBonos:

    def __init__(self):
        self.empleados = []

    def agregar_empleado(self, empleado):
        self.empleados.append(empleado)

    def calcular_presupuesto_total(self):
        total = 0

        for empleado in self.empleados:
            bono_mensual = empleado.calcular_bono()
            total += bono_mensual * 3

        return total

    def generar_informe(self):

        if not self.empleados:
            print("No hay empleados registrados en el sistema")
            return

        total_empleados = len(self.empleados)
        total_dependientes = 0
        suma_salarios = 0
        suma_antiguedad = 0
        suma_bonos = 0

        for empleado in self.empleados:
            total_dependientes += len(empleado.dependientes)
            suma_salarios += empleado.salario
            suma_antiguedad += empleado.antiguedad
            suma_bonos += empleado.calcular_bono()

        promedio_salario = suma_salarios / total_empleados
        promedio_antiguedad = suma_antiguedad / total_empleados
        promedio_bono = suma_bonos / total_empleados

        presupuesto_total = self.calcular_presupuesto_total()


        print("\nINFORME DEL SISTEMA DE BONOS")
        print("=" * 40)
        print(f"Total de empleados: {total_empleados}")
        print(f"Total de dependientes: {total_dependientes}")
        print(f"Salario promedio: ${promedio_salario:.2f}")
        print(f"Antigüedad promedio: {promedio_antiguedad:.1f} años")
        print(f"Bono promedio: ${promedio_bono:.2f}")
        print(f"Presupuesto total (3 meses): ${presupuesto_total:.2f}")

def crear_ejemplos():
    sistema = SistemaBonos()

    # Ejemplo 1: Un médico con larga trayectoria y dos hijos estudiantes
    medico = Empleado("Juan Pérez", 150000, 18)

    hijo1 = Dependiente("Lucas Pérez", 2010, True)  # Hijo de 15 años que estudia
    hijo2 = Dependiente("Ana Pérez", 2015, True)    # Hija de 10 años que estudia

    medico.agregar_dependiente(hijo1)
    medico.agregar_dependiente(hijo2)

    # Ejemplo 2: Una enfermera con poca antigüedad y sin hijos
    enfermera = Empleado("María López", 80000, 3)

    # Ejemplo 3: Un técnico con antigüedad media y un hijo que no estudia
    tecnico = Empleado("Carlos Gómez", 70000, 8)
    hijo_tecnico = Dependiente("Pablo Gómez", 2020, False)  # Hijo de 5 años que no estudia
    tecnico.agregar_dependiente(hijo_tecnico)

    sistema.agregar_empleado(medico)
    sistema.agregar_empleado(enfermera)
    sistema.agregar_empleado(tecnico)

    return sistema


def main():

    sistema = crear_ejemplos()

    print("LISTA DE EMPLEADOS Y SUS DEPENDIENTES:")
    for i, empleado in enumerate(sistema.empleados, 1):
        print(f"\n{i}. {empleado}")

        if empleado.dependientes:
            print("   Dependientes:")
            for j, dependiente in enumerate(empleado.dependientes, 1):
                print(f"   {j}. {dependiente}")

    sistema.empleados[0].mostrar_detalles_bono()

    sistema.generar_informe()


if __name__ == "__main__":
    main()
```

## **Ejercicio 8:** Flota de Vehículos

Diseña un sistema simplificado para gestionar una flota de vehículos. Crea una clase Vehiculo con atributos como marca, modelo, año, placa (verificar que siempre sea única) y tipo (que puede ser "coche", "camioneta", "motocicleta", etc.), y métodos para mostrar su información y actualizar su estado (por ejemplo, "disponible", "en reparación", "alquilado").  Luego, crea una clase Flota que contenga un diccionario donde las claves sean las placas y los valores sean objetos Vehiculo. La clase Flota debe tener métodos para agregar vehículos (verificando que no haya placas duplicadas), eliminar vehículos, buscar vehículos por placa, listar todos los vehículos de un tipo específico (usando una lista), y calcular el promedio de antigüedad de todos los vehículos en la flota; adicionalmente, implementa un método que retorne una tupla indicando la marca y el modelo del vehículo más antiguo.

```python
import datetime

class Vehiculo:
    """
    Clase que representa un vehículo en una flota.
    """

    def __init__(self, marca, modelo, anio, placa, tipo):
        """
        Inicializa un nuevo vehículo.
        """
        self.marca = marca
        self.modelo = modelo
        self.anio = anio
        self.placa = placa
        self.tipo = tipo
        self.estado = "disponible"  # Estado inicial: disponible

    def __str__(self):
        """
        Devuelve una representación en string del vehículo.
        """
        return f"{self.marca} {self.modelo} ({self.anio}) - Placa: {self.placa}, Tipo: {self.tipo}, Estado: {self.estado}"

    def mostrar_informacion(self):
        """
        Muestra la información completa del vehículo.
        """
        return f"""
        Información del Vehículo:
        ------------------------
        Marca: {self.marca}
        Modelo: {self.modelo}
        Año: {self.anio}
        Placa: {self.placa}
        Tipo: {self.tipo}
        Estado actual: {self.estado}
        """

    def actualizar_estado(self, nuevo_estado):
        """
        Actualiza el estado del vehículo.
        """
        estados_validos = ["disponible", "en reparación", "alquilado", "en mantenimiento", "fuera de servicio"]

        if nuevo_estado.lower() in estados_validos:
            self.estado = nuevo_estado.lower()
            return True
        else:
            print(f"Estado '{nuevo_estado}' no válido. Estados válidos: {', '.join(estados_validos)}")
            return False


class Flota:
    """
    Clase que gestiona una flota de vehículos.
    """

    def __init__(self, nombre="Flota General"):
        """
        Inicializa una nueva flota.
        """
        self.nombre = nombre
        self.vehiculos = {}  # Diccionario con placas como claves y objetos Vehiculo como valores

    def agregar_vehiculo(self, vehiculo):
        """
        Agrega un vehículo a la flota verificando que no exista ya la placa.
        """
        if vehiculo.placa in self.vehiculos:
            print(f"Error: Ya existe un vehículo con la placa {vehiculo.placa}")
            return False

        self.vehiculos[vehiculo.placa] = vehiculo
        print(f"Vehículo con placa {vehiculo.placa} agregado correctamente.")
        return True

    def eliminar_vehiculo(self, placa):
        """
        Elimina un vehículo de la flota por su placa.
        """
        if placa in self.vehiculos:
            del self.vehiculos[placa]
            print(f"Vehículo con placa {placa} eliminado correctamente.")
            return True
        else:
            print(f"Error: No se encontró ningún vehículo con la placa {placa}")
            return False

    def buscar_vehiculo(self, placa):
        """
        Busca un vehículo por su placa.
        """
        return self.vehiculos.get(placa, None)

    def listar_vehiculos_por_tipo(self, tipo):
        """
        Lista todos los vehículos de un tipo específico.
        """
        return [vehiculo for vehiculo in self.vehiculos.values() if vehiculo.tipo.lower() == tipo.lower()]

    def calcular_promedio_antiguedad(self):
        """
        Calcula el promedio de antigüedad de todos los vehículos en la flota.
        """
        if not self.vehiculos:
            return 0

        anio_actual = datetime.datetime.now().year
        suma_edades = sum(anio_actual - vehiculo.anio for vehiculo in self.vehiculos.values())

        return suma_edades / len(self.vehiculos)

    def obtener_vehiculo_mas_antiguo(self):
        """
        Obtiene el vehículo más antiguo de la flota.
        """
        if not self.vehiculos:
            return None

        vehiculo_mas_antiguo = min(self.vehiculos.values(), key=lambda v: v.anio)
        return (vehiculo_mas_antiguo.marca, vehiculo_mas_antiguo.modelo)

    def mostrar_todos_vehiculos(self):
        """
        Muestra todos los vehículos de la flota.
        """
        if not self.vehiculos:
            return "La flota no tiene vehículos registrados."

        resultado = f"Vehículos en la flota '{self.nombre}':\n"
        for i, vehiculo in enumerate(self.vehiculos.values(), 1):
            resultado += f"{i}. {str(vehiculo)}\n"

        return resultado

def main():
    # Crear una flota
    mi_flota = Flota("Flota Concordia")

    # Agregar algunos vehículos
    vehiculo1 = Vehiculo("Toyota", "Corolla", 2020, "1235AKX", "coche")
    vehiculo2 = Vehiculo("Honda", "Civic", 2019, "7898FGH", "coche")
    vehiculo3 = Vehiculo("Ford", "Ranger", 2015, "4565YTGH", "camioneta")
    vehiculo4 = Vehiculo("Yamaha", "MT-07", 2022, "9079YTH", "motocicleta")
    vehiculo5 = Vehiculo("Chevrolet", "S10", 2010, "1112RET", "camioneta")

    mi_flota.agregar_vehiculo(vehiculo1)
    mi_flota.agregar_vehiculo(vehiculo2)
    mi_flota.agregar_vehiculo(vehiculo3)
    mi_flota.agregar_vehiculo(vehiculo4)
    mi_flota.agregar_vehiculo(vehiculo5)

    # Intentar agregar un vehículo con placa duplicada
    vehiculo_duplicado = Vehiculo("Nissan", "Sentra", 2021, "1235AKX", "coche")
    mi_flota.agregar_vehiculo(vehiculo_duplicado)

    # Mostrar todos los vehículos
    print("\n" + mi_flota.mostrar_todos_vehiculos())

    # Buscar un vehículo por placa
    placa_buscar = "4565YTGH"
    vehiculo_encontrado = mi_flota.buscar_vehiculo(placa_buscar)
    if vehiculo_encontrado:
        print(f"\nVehículo encontrado con placa {placa_buscar}:")
        print(vehiculo_encontrado.mostrar_informacion())

    # Actualizar estado de un vehículo
    vehiculo_encontrado.actualizar_estado("en reparación")
    print(f"Estado actualizado: {vehiculo_encontrado}")


    # Listar vehículos por tipo
    tipo_buscar = "camioneta"
    camionetas = mi_flota.listar_vehiculos_por_tipo(tipo_buscar)
    print(f"\nVehículos de tipo '{tipo_buscar}':")
    for i, camioneta in enumerate(camionetas, 1):
        print(f"{i}. {camioneta}")

    # Calcular promedio de antigüedad
    promedio = mi_flota.calcular_promedio_antiguedad()
    print(f"\nPromedio de antigüedad de la flota: {promedio:.2f} años")

    # Obtener vehículo más antiguo
    mas_antiguo = mi_flota.obtener_vehiculo_mas_antiguo()
    print(f"\nEl vehículo más antiguo es: {mas_antiguo[0]} {mas_antiguo[1]}")

    # Eliminar un vehículo
    mi_flota.eliminar_vehiculo("1112RET")

    # Mostrar la flota actualizada
    print("\n" + mi_flota.mostrar_todos_vehiculos())


if __name__ == "__main__":
    main()
```
