### Crear un diccionario que contenga un registro por cada imagen/archivo en el dataset. El registro tendra que estar compuesto por ruta, categoría, alto, y ancho.

```python
# Crear un diccionario con ruta, categoría, alto y ancho
from PIL import Image
import os

base_path = "./cat-vs-rabbit/train-cat-rabbit"
categorias = ["cat", "rabbit"]
datos_imagenes = []
dato_imagen = {}

for categoria in categorias:
    carpeta = os.path.join(base_path, categoria)
    for archivo in os.listdir(carpeta):
        if archivo.endswith(('.jpg', '.jpeg', '.png')):
            ruta = os.path.join(carpeta, archivo)
            try:
                with Image.open(ruta) as img:
                    ancho, alto = img.size
                    dato_imagen = {
                        "ruta": ruta,
                        "categoria": categoria,
                        "alto": alto,
                        "ancho": ancho
                    }
                datos_imagenes.append(dato_imagen)
            except:
                print(f"Error con la imagen: {ruta}")
print(datos_imagenes)
```

### Realizar un análisis sobre las composición de las etiquetas presentes en el dataset.

```python
from collections import Counter
conteo_etiquetas = Counter([img['categoria'] for img in datos_imagenes])
#print(Counter('abcdeabcdabcaba'))
print("\nConteo por categoría:")
print(conteo_etiquetas)
```

### Mostrar un ejemplo por tipo de objeto o categoría.


```python
import random
import matplotlib.pyplot as plt
def mostrar_ejemplo(categoria):
    for img in datos_imagenes:
      random_index = random.randint(0,len(datos_imagenes)-1)
      img = datos_imagenes[random_index]
      if img['categoria'] == categoria:
          imagen = Image.open(img['ruta'])
          #plt.subplot(3, 3, 1)
          plt.imshow(imagen)
          plt.title(f"Ejemplo: {categoria}")
          plt.axis('off')
          plt.show()
          break

mostrar_ejemplo("cat")
mostrar_ejemplo("rabbit")
```

### Generar un gráfico con la distribución de tamaños (alto y ancho de las imágenes).


```python
# Paso 8: Gráfico con la distribución de tamaños
import seaborn as sns
import pandas as pd

# Convertimos a DataFrame para graficar
df = pd.DataFrame(datos_imagenes)
sns.jointplot(data=df, x="ancho", y="alto", hue="categoria", kind="scatter")
plt.suptitle("Distribución de Tamaños de Imágenes")
plt.show()
```

### Mostrar cual es la categoría que mas y menos se repiten.

```python
categoria_mas = conteo_etiquetas.most_common(1)[0]
categoria_menos = conteo_etiquetas.most_common()[-1]
print(f"\nCategoría que más se repite: {categoria_mas[0]} ({categoria_mas[1]} veces)")
print(f"Categoría que menos se repite: {categoria_menos[0]} ({categoria_menos[1]} veces)")
```
